```python
import pandas as pd
import numpy as np

from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
def getHTMLcontent(link):
    html = urlopen(link)
    soup = BeautifulSoup(html, 'html.parser')
    return soup
```

```python
content = getHTMLcontent('https://en.wikipedia.org/wiki/Tourist_attractions_in_West_Bengal')
tables = content.find_all('table')

for table in tables:
    print(table.prettify())
```

```python
table = content.find('table',{'class':'wikitable'})
rows = table.find_all('tr')

for row in rows:
    cells = row.find_all('td')
    if len(cells) > 1:
        places_link = cells[0].find('a')
        print(places_link.get('href'))
```

```python
def getAdditionalDetails(url):
    location = getHTMLcontent('https://en.wikipedia.org' + url)
    try:
        latitude = location.find('span',{'class':'latitude'})
        longitude = location.find('span',{'class':'longitude'})
        return [latitude.get_text(),longitude.get_text()]
    except Exception as error:
        return [-1,-1]
```

```python
data_content = []
for row in rows:
    cells = row.find_all('td')
    if len(cells) > 1:
        location_info = [cell.text.strip('\n') for cell in cells]
        location_link = cells[0].find('a')
        location_info += getAdditionalDetails(location_link.get('href'))
        data_content.append(location_info)

dataset = pd.DataFrame(data_content)
headers = ['Tourist Spots','Location','','Latitude','Longitude']
dataset.columns = headers
dataset.to_csv('Dataset_Project.csv', index = False)
```

```python
def convertLattoDecimalLat(Lat):
    if(len(Lat)>7):
        Latitude = float(Lat[0:Lat.index("")]) + (float(Lat[Lat.index("")+1:Lat.index("")])/60) + (float(Lat[Lat.index("")+1:Lat.index("")])/3600)
    else:
        Latitude = float(Lat[0:Lat.index("")]) + (float(Lat[Lat.index("")+1:Lat.index("")])/60)
    return Latitude
def convertLongtoDecimalLong(Long):
    if(len(Long)>7):
        Longitude = float(Long[0:Long.index("")]) + (float(Long[Long.index("")+1:Long.index("")])/60) + (float(Long[Long.index("")+1:Long.index("")])/3600)
    else:
        Longitude = float(Long[0:Long.index("")]) + (float(Long[Long.index("")+1:Long.index("")])/60)
    return Longitude
```
